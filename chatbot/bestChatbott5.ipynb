```python
from transformers import T5ForConditionalGeneration, T5Tokenizer, AdamW
from torch.utils.data import DataLoader, Dataset
import wandb
import torch

class EwhaQADataset(Dataset):
    def __init__(self, data, tokenizer, max_length):
        self.data = data
        self.tokenizer = tokenizer
        self.max_length = max_length

    def __len__(self):
        return len(self.data)

    def __getitem__(self, index):
        item = self.data[index]
        question = item["question"]
        answer = item["answer"]

        encoding = self.tokenizer.encode_plus(
            question,
            answer,
            add_special_tokens=True,
            max_length=self.max_length,
            padding="max_length",
            truncation=True,
            return_tensors="pt"
        )

        input_ids = encoding["input_ids"].squeeze()
        attention_mask = encoding["attention_mask"].squeeze()

        return {
            "input_ids": input_ids,
            "attention_mask": attention_mask,
        }

wandb.init(project="chatbot-training", name="t5-base-fine-tuning")

def fine_tune_model(train_data):
    tokenizer = T5Tokenizer.from_pretrained("t5-base")
    model = T5ForConditionalGeneration.from_pretrained("t5-base")
    
    # Set up training parameters
    max_length = 128
    batch_size = 8
    num_epochs = 35

    # Create dataset and dataloader for training
    dataset = EwhaQADataset(train_data, tokenizer, max_length)
    dataloader = DataLoader(dataset, batch_size=batch_size, shuffle=True)

    # Set up optimizer
    optimizer = torch.optim.AdamW(model.parameters(), lr=1e-5)

    model.train()
    for epoch in range(num_epochs):
        total_train_loss = 0.0

        for batch in dataloader:
            input_ids = batch["input_ids"]
            attention_mask = batch["attention_mask"]

            # Forward pass
            outputs = model(
                input_ids=input_ids,
                attention_mask=attention_mask,
                labels=input_ids
            )

            loss = outputs.loss
            loss.backward()

            # Update model weights
            optimizer.step()
            optimizer.zero_grad()
            total_train_loss += loss.item()
            
        avg_train_loss = total_train_loss / len(train_data)

        # Log losses to wandb
        wandb.log({"epoch": epoch+1, "train_loss": avg_train_loss})

        # Print training progress
        print(f"Epoch {epoch+1}/{num_epochs}, Train Loss: {avg_train_loss}")
        
    model.save_pretrained("fine_tuned_model")
    tokenizer.save_pretrained("fine_tuned_model")

# Call the fine-tuning function
fine_tune_model(training_dataset)

# Finish wandb run
wandb.finish()
```
